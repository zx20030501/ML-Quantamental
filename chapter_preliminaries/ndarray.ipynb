## è¿ç®—ç¬¦

æˆ‘ä»¬çš„å…´è¶£ä¸ä»…é™äºè¯»å–æ•°æ®å’Œå†™å…¥æ•°æ®ã€‚
æˆ‘ä»¬æƒ³åœ¨è¿™äº›æ•°æ®ä¸Šæ‰§è¡Œæ•°å­¦è¿ç®—ï¼Œå…¶ä¸­æœ€ç®€å•ä¸”æœ€æœ‰ç”¨çš„æ“ä½œæ˜¯*æŒ‰å…ƒç´ *ï¼ˆelementwiseï¼‰è¿ç®—ã€‚
å®ƒä»¬å°†æ ‡å‡†æ ‡é‡è¿ç®—ç¬¦åº”ç”¨äºæ•°ç»„çš„æ¯ä¸ªå…ƒç´ ã€‚
å¯¹äºå°†ä¸¤ä¸ªæ•°ç»„ä½œä¸ºè¾“å…¥çš„å‡½æ•°ï¼ŒæŒ‰å…ƒç´ è¿ç®—å°†äºŒå…ƒè¿ç®—ç¬¦åº”ç”¨äºä¸¤ä¸ªæ•°ç»„ä¸­çš„æ¯å¯¹ä½ç½®å¯¹åº”çš„å…ƒç´ ã€‚

åœ¨æ•°å­¦è¡¨ç¤ºæ³•ä¸­ï¼Œæˆ‘ä»¬å°†é€šè¿‡ç¬¦å·$f: \mathbb{R} \rightarrow \mathbb{R}$

è¡¨ç¤º*äºŒå…ƒ*æ ‡é‡è¿ç®—ç¬¦ï¼Œè¿™æ„å‘³ç€è¯¥å‡½æ•°æ¥æ”¶ä¸¤ä¸ªè¾“å…¥ï¼Œå¹¶äº§ç”Ÿä¸€ä¸ªè¾“å‡ºã€‚
ç»™å®šåŒä¸€å½¢çŠ¶çš„ä»»æ„ä¸¤ä¸ªå‘é‡$\mathbf{u}$å’Œ$\mathbf{v}$å’ŒäºŒå…ƒè¿ç®—ç¬¦$f$ï¼Œ
