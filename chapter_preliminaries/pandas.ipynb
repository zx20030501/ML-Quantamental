è¦[**ä»åˆ›å»ºçš„CSVæ–‡ä»¶ä¸­åŠ è½½åŸå§‹æ•°æ®é›†**]ï¼Œæˆ‘ä»¬å¯¼å…¥`pandas`åŒ…å¹¶è°ƒç”¨`read_csv`å‡½æ•°ã€‚è¯¥æ•°æ®é›†æœ‰å››è¡Œä¸‰åˆ—ã€‚å…¶ä¸­æ¯è¡Œæè¿°äº†æˆ¿é—´æ•°é‡ï¼ˆâ€œNumRoomsâ€ï¼‰ã€å··å­ç±»å‹ï¼ˆâ€œAlleyâ€ï¼‰å’Œæˆ¿å±‹ä»·æ ¼ï¼ˆâ€œPriceâ€ï¼‰ã€‚

## å¤„ç†ç¼ºå¤±å€¼

æ³¨æ„ï¼Œâ€œNaNâ€é¡¹ä»£è¡¨ç¼ºå¤±å€¼ã€‚
[**ä¸ºäº†å¤„ç†ç¼ºå¤±çš„æ•°æ®ï¼Œå…¸å‹çš„æ–¹æ³•åŒ…æ‹¬*æ’å€¼æ³•*å’Œ*åˆ é™¤æ³•*ï¼Œ**]
å…¶ä¸­æ’å€¼æ³•ç”¨ä¸€ä¸ªæ›¿ä»£å€¼å¼¥è¡¥ç¼ºå¤±å€¼ï¼Œè€Œåˆ é™¤æ³•åˆ™ç›´æ¥å¿½ç•¥ç¼ºå¤±å€¼ã€‚

å…¶ä¸­å‰è€…ä¸º`data`çš„å‰ä¸¤åˆ—ï¼Œè€Œåè€…ä¸º`data`çš„æœ€åä¸€åˆ—ã€‚
å¯¹äº`inputs`ä¸­ç¼ºå°‘çš„æ•°å€¼ï¼Œæˆ‘ä»¬ç”¨åŒä¸€åˆ—çš„å‡å€¼æ›¿æ¢â€œNaNâ€é¡¹ã€‚


````python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean())
print(inputs)
````

[**å¯¹äº`inputs`ä¸­çš„ç±»åˆ«å€¼æˆ–ç¦»æ•£å€¼ï¼Œæˆ‘ä»¬å°†â€œNaNâ€è§†ä¸ºä¸€ä¸ªç±»åˆ«ã€‚**]
ç”±äºâ€œå··å­ç±»å‹â€ï¼ˆâ€œAlleyâ€ï¼‰åˆ—åªæ¥å—ä¸¤ç§ç±»å‹çš„ç±»åˆ«å€¼â€œPaveâ€å’Œâ€œNaNâ€ï¼Œ
`pandas`å¯ä»¥è‡ªåŠ¨å°†æ­¤åˆ—è½¬æ¢ä¸ºä¸¤åˆ—â€œAlley_Paveâ€å’Œâ€œAlley_nanâ€ã€‚
å··å­ç±»å‹ä¸ºâ€œPaveâ€çš„è¡Œä¼šå°†â€œAlley_Paveâ€çš„å€¼è®¾ç½®ä¸º1ï¼Œâ€œAlley_nanâ€çš„å€¼è®¾ç½®ä¸º0ã€‚
ç¼ºå°‘å··å­ç±»å‹çš„è¡Œä¼šå°†â€œAlley_Paveâ€å’Œâ€œAlley_nanâ€åˆ†åˆ«è®¾ç½®ä¸º0å’Œ1ã€‚


````python
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
````
